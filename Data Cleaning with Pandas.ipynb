```python
import pandas as pd
pd.set_option("display.max_colwidth", None)
# 0 is always missing data
df = pd.read_csv("neiss2017_edited.tsv",
                 sep="\t",
                na_values= [0],
                encoding="latin-1")
df.head(20)
```

```python
df.Age
```

```python
# rip the YO off
# can't do .astype(int)
# must do .astype(float) for missing data
df.Age = df.Age.str.replace("YO", "").astype(float)
```

```python
df[df['Age'] >= 200]
```

```python
# get rid of the babies
df = df[df.Age < 200]
df.Age.hist()
```

```python
# df[(df.Other_Race == "HISPANIC") | (df.Other_Race == 'HISP')]
df[df.Other_Race.str.contains("^HISP", na = False)].Other_Race.value_counts()
```

```python
df.loc[df.Other_Race.str.contains("^HISP", na = False)]
df.loc[df.Other_Race.str.contains("^HISP", na = False), "Other_Race"] = "HISPANIC"
# use df.loc[<<<condition>>>, 'column name'], 'value'
# to replace the data that fits the criteria in certain column with certain value
```

```python
df.Other_Race = df.Other_Race.replace({
    "MULTIRACIAL": "MULTI-RACIAL",
    "MULTI RACIAL": "MULTI-RACIAL"
})
df.Other_Race.value_counts().head(20)
```

```python
df['Narrative'] = df.Narrative_1 + df.Narrative_2.fillna("")
df.head()
```

```python
df[df.Narrative.str.contains("PUNCH.* WALL")]
```

```python
df.Narrative.str.extract("PUNCHED A (\w+)", expand = False).value_counts()
# \w+ = match a word
# \d+ = match a number
# (\w\s)+ = match word characters or white spaces
```

```python
# CONDITIONAL ROW UPDATES!!!! FOR A SPECIFIC COLUMN!!!!!
# For all of these people I want to update the 'Other Race' column
# use df.loc[<<<condition>>>, 'colname'] = 'value'
df.loc[df.Other_Race.str.contains("^HISP", na=False)]
df.loc[df.Other_Race.str.contains("^HISP", na=False), "Other_Race"] = "HISPANIC"
```
